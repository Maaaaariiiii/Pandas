# Redundant columns identification and elimination

```python
# Calculate the variance for each column
variance = numerical_columns.var()

# Find columns with zero or very low variance (you can set a threshold)
redundant_columns = variance[variance <= 0.01]

# Print redundant columns
print("Redundant Columns:")
print(redundant_columns)
```

```python
threshold = 0.98
redundant_columns_2 = [i for i in categorical_columns.columns if categorical_columns[i].value_counts(normalize=True).max() >= threshold]

# Print constant columns
print("Ineffective Columns:")
print(redundant_columns_2)

```

Drpping redundant columns from dataset

```python
data= data.drop(columns=['EmployeeCount', 'StandardHours', 'Over18'])
```

```python
data.columns
```

Renaming the columns to user friendly names

```python
data= data.rename(columns={
    
    'Age': 'Age',
    'DailyRate': 'Daily_Rate',
    'DistanceFromHome': 'Distance_From_Home',
    'EmployeeNumber': 'Employee_ID','EducationField':'Education_Field',
    'EnvironmentSatisfaction': 'Environment_Satisfaction_Level',
    'HourlyRate': 'Hourly_Rate',
    'JobInvolvement': 'Job_Involvement',
    'JobSatisfaction': 'Job_Satisfaction_Level',
    'MonthlyIncome': 'Monthly_Income',
    'MonthlyRate': 'Monthly_Rate',
    'NumCompaniesWorked': 'No_Companies_Worked',
    'PercentSalaryHike': 'Percentage_Salary_Hike',
    'PerformanceRating': 'Performance_Rating',
    'RelationshipSatisfaction': 'Relationship_Satisfaction_Level',
    'StockOptionLevel': 'Stock_Option_Level',
    'TotalWorkingYears': 'Total_Working_Years',
    'TrainingTimesLastYear': 'Training_No_LastYear',
    'WorkLifeBalance': 'Work_Life_Balance_Level',
    'YearsAtCompany': 'Years_At_Company',
    'YearsInCurrentRole': 'Years_In_Current_Role',
    'YearsSinceLastPromotion': 'Years_Since_LastPromotion',
    'YearsWithCurrManager': 'Years_With_Current_Manager',
    'Attrition': 'Attrition',
    'BusinessTravel': 'Business_Travel',
    'OverTime': 'Over_Time'

                                 })
```

Checking duplicate items for employee_ID

```python
duplicate_employee_ids = data[data.duplicated(subset=['Employee_ID'], keep=False)]

grouped_duplicates = duplicate_employee_ids.groupby('Employee_ID')


for employee_id, duplicates in grouped_duplicates:
    print(f"Duplicate entries found for Employee ID {employee_id}:")
    print(duplicates)
    
```

```python
data=data.drop_duplicates()
```

Remove the NaN values from the dataset


```python
data=data.dropna()
```

Feature Engineering for Age feature to make easy to interprete results based on Age

```python
data['AgeIntervals']=pd.cut(data['Age'], bins =4,labels=['Group1',
                                        'Group2',
                                         'Group3',
                                         'Group4',
                                           ],
                                         include_lowest=True)
```
